```python
# Libraries
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from math import pi
import seaborn as sns
 
# Set data
df = pd.DataFrame({
'group': ['A','B','C','D','A'],
'var1': [38, 1.5, 30, 4,3],
'var2': [29, 10, 9, 34,3],
'var3': [8, 39, 23, 24,12],
'var4': [7, 31, 33, 14,23],
'var5': [28, 15, 32, 14,34],
'var6': [12,10,5,12,34]
})
```

```python
df
```

```python
labels=np.array(['var1', 'var2', 'var3', 'var4', 'var5', 'var6'])
stats=df.loc[0,labels].values

print(labels)
print(stats)
```

```python
angles=np.linspace(0, 2*np.pi, len(labels), endpoint=False)

# close the plot
stats=np.concatenate((stats,[stats[0]]))
angles=np.concatenate((angles,[angles[0]]))

print(stats)
print(angles)
```

```python
fig=plt.show()
ax = plt.subplot(111, polar=True)
ax.plot(angles, stats, 'o-', linewidth=2)
ax.fill(angles, stats, alpha=0.25)
ax.set_thetagrids((angles * 180/np.pi)[0:6], labels)
ax.set_title([df.loc[0,"group"]])
ax.grid(True)
```
